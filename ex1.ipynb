## Data download and processing Helper Code

````python
def download_and_extract(url, download_dir):
    """
    Download and extract the CIFAR-10 dataset if it doesn't already exist.

    Parameters
    ----------
    url : str
        Internet URL for the tar-file to download.
        Example: "https://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz"
    download_dir : str
        Directory where the downloaded file will be saved and extracted.
        Example: "data/CIFAR-10/"

    Returns
    -------
    None
    """

    # Determine filename and full path where the file will be saved
    filename = url.split('/')[-1]
    file_path = os.path.join(download_dir, filename)

    # Check if dataset is already downloaded (and extracted)
    if not os.path.exists(file_path):
       # Create the directory if it doesn’t exist
        if not os.path.exists(download_dir):
            os.makedirs(download_dir)

        # Download the dataset
        print("Downloading, This might take several minutes.")
        last_update_time = time.time()
        file_path, _ = urllib.request.urlretrieve(url=url, filename=file_path)

        print()
        print("Download finished. Extracting files.")

        # Extract the dataset
        if file_path.endswith(".zip"):
            # Unpack the zip-file.
            zipfile.ZipFile(file=file_path, mode="r").extractall(download_dir)
        elif file_path.endswith((".tar.gz", ".tgz")):
            # Unpack the tar-ball.
            tarfile.open(name=file_path, mode="r:gz").extractall(download_dir)

        print("Done. Dataset is ready!")
    else:
        print("Dataset already downloaded and unpacked.")
        print("If something seems wrong, delete the folder and re-run.")


def load_CIFAR_batch(filename):
    ''' Load a single batch of the CIFAR-10 dataset.'''
    with open(filename, 'rb') as f:
        datadict = pickle.load(f, encoding = 'latin1')
        X = datadict['data']
        Y = datadict['labels']

        # Reshape and transpose: original shape (10000, 3072)
        X = X.reshape(10000, 3, 32, 32).transpose(0, 2, 3, 1).astype("float")
        Y = np.array(Y)
        return X, Y


def load(ROOT):
    ''' Load all training and test batches of CIFAR-10.'''
    xs = []
    ys = []
    for b in range(1, 6):
        f = os.path.join(ROOT, 'data_batch_%d' % (b, ))
        X, Y = load_CIFAR_batch(f)
        xs.append(X)
        ys.append(Y)
    Xtr = np.concatenate(xs)
    Ytr = np.concatenate(ys)
    del X, Y
    Xte, Yte = load_CIFAR_batch(os.path.join(ROOT, 'test_batch'))
    return Xtr, Ytr, Xte, Yte
````

## Data Preprocessing Part 2

In this section, we apply several image preprocessing steps to prepare the data for our linear models:

1. Mean Subtraction - We subtract the mean image (computed over the training set) from every image.

2. Flattening - Each image originally has the shape (32 × 32 × 3).
We flatten it into a single vector of 3072 features.

3. Bias Trick - We append an additional constant feature (set to 1) to every image vector.

````python
# =========================
# Final data preprocessing
# =========================

# 1) center + standardize using TRAIN statistics (feature-wise)
mean_image = np.mean(X_train, axis=0, keepdims=True)
std_image  = np.std(X_train,  axis=0, keepdims=True) + 1e-8
X_train = (X_train - mean_image) / std_image
X_val   = (X_val   - mean_image) / std_image
X_test  = (X_test  - mean_image) / std_image

# 2) flatten HxWxC -> D
X_train = X_train.reshape(X_train.shape[0], -1)
X_val   = X_val.reshape(X_val.shape[0], -1)
X_test  = X_test.reshape(X_test.shape[0], -1)

# 3) add a bias term (last feature = 1)
X_train = np.hstack([X_train, np.ones((X_train.shape[0], 1))])
X_val   = np.hstack([X_val,   np.ones((X_val.shape[0],   1))])
X_test  = np.hstack([X_test,  np.ones((X_test.shape[0],  1))])

print(f"Shape of Training Set: {X_train.shape}")   # (N_train, D+1)
print(f"Shape of Validation Set: {X_val.shape}")   # (N_val,   D+1)
print(f"Shape of Test Set: {X_test.shape}")        # (N_test,  D+1)

# Ensure labels are 0..C-1
num_classes = int(np.max(y_train)) + 1
assert set(np.unique(y_train)) <= set(range(num_classes)), "y must be in 0..C-1"
````

### ❀ Q2: Understanding the Image Preprocessing ❀

Before moving on, let’s review the preprocessing steps applied to our dataset.

Answer the following questions:

1. Mean Centering & Standardization

- Why do we subtract the mean image AND divide by the per-feature standard deviation (both computed from the training set) from every image?
- Why is it important to use the *training set's* statistics (mean and std) when transforming the validation and test sets, rather than computing fresh statistics on each split?

2. Flattening

- Why do we need to flatten the images into 3072-dimensional vectors before training a linear model?


3. Bias Trick

- Why do we add a constant 1 to every image vector (known as the bias trick)?

**(5 points)**



**Answers**

**1. Mean centering & standardization**

- We subtract the mean image and divide by the per-feature standard deviation so every pixel has roughly zero mean and unit variance. This puts all features on the same numerical scale, which (a) keeps gradients well-conditioned so a single global learning rate works across features, (b) accelerates convergence of SGD, and (c) prevents very bright or very dark pixel coordinates from dominating the linear score $s = Wx$. Without centering, the bias column has to absorb the mean, which slows training; without scaling, features with larger variance push the gradient direction toward themselves.

- We must use the *training* statistics on the validation and test sets because these sets are meant to simulate unseen data. Re-computing the mean/std on each split would (1) leak information from the validation/test set into preprocessing, which biases evaluation upward, and (2) apply a different transformation at test time than at train time, meaning the weights we learned are no longer applied to comparable inputs. In production, we never see the true test distribution ahead of time, so we must freeze the transform using only what is visible during training.

**2. Flattening**

- A linear classifier computes $s = Wx + b$ and expects each example as a single feature vector. A CIFAR-10 image is a $32 \times 32 \times 3$ tensor, but the model does not exploit its 2-D spatial structure (that is what CNNs are for). Flattening to a 3072-dimensional vector simply reshapes the same numbers into the form the matrix multiplication needs. The spatial structure is lost, which is a real limitation of linear models on images — but it is consistent with what the model can actually use.

**3. Bias trick**

- Appending a constant $1$ to every input vector lets us fold the bias $b$ into an extra row of $W$, so the forward pass becomes a single matrix multiply $s = W^\top x$ instead of $s = W^\top x + b$. This simplifies both the code and the math (e.g. the gradient has a single tensor $dW$ instead of separate $dW, db$), and is numerically equivalent to keeping an explicit bias term.

## Linear perceptron

Our first linear classifier is the **Perceptron**, a foundational model in machine learning.

It computes a **linear score** for each class:

$$ f(x_i; W, b) = W \cdot x_i + b $$

- W : a weight matrix of shape *(num_features × num_classes)*  
- b : a bias term (included using the **bias trick**)

Each column in \( W \) corresponds to one class.  
The **final prediction** is the class with the **highest score**.

---

### Multiclass vs. Binary

- **Multiclass Case:**  
  `W` is a matrix. Each row corresponds to a feature, each column to a class.  
  Prediction is made via:

  $$ \hat{y_i} = \arg\max_j (W_j \cdot x_i) $$

- **Binary Case (special case of 2 classes):**  
  You can either:
  1. Treat it as a **2-class multiclass** problem (use 2 columns in `W`),  
     prediction = class with the highest score.  
  2. Or use a **single weight vector** `w`, where:
    $$ \text{predict } 1 \text{ if } w \cdot x + b > 0, \text{ else } 0 $$

In this exercise, our implementation is **multiclass**, supporting more than 2 classes.


### Objective

Your goal is to **learn** the parameters `W` and `b` such that the classifier correctly predicts the labels in the training set.  We’ll do this by **minimizing the Perceptron loss**, which penalizes misclassified examples.

## Logistic regression

Logistic regression can be generalized from binary classification to **multiclass** by applying the **softmax** function to the linear scores:

$$
\text{scores}(x) = XW \quad,\quad
p(y=j \mid x) = \frac{\exp(\text{scores}_j)}{\sum_k \exp(\text{scores}_k)}
$$


- `W` : weight matrix (one column per class).
- The model outputs a **probability distribution** over classes.
- We train by minimizing **categorical cross-entropy** (a.k.a. softmax loss).

**Numerical stability:** Always subtract the row-wise max before exponentiating (log-sum-exp trick).

## Multiclass Cross-Entropy Loss (Softmax)

In this section, you’ll implement the **softmax cross-entropy loss**, also known as the **multiclass logistic loss**.

This loss generalizes binary cross-entropy to multiple classes.  
Given scores $s = XW$ for each class, we first normalize them into probabilities using **softmax**:


$$ p(y=j|x_i) = \frac{e^{s_j}}{\sum_{k} e^{s_k}} $$

The loss for a single example is the **negative log-likelihood** of the correct class:


$$ L_i = -\log(p(y_i|x_i))$$

The overall loss is the average across all examples.

### ❀ Q4: Error Analysis — Confusion Matrix ❀

Using the best logistic-regression model from your hyperparameter sweep, compute the 3×3 confusion matrix on the test set. You may use `np.add.at` or a double loop — do **not** use `sklearn`.

1. Print the confusion matrix with clear row/column labels (rows = true class, columns = predicted class).
2. Which pair of classes is confused most often? Does that match what you would expect from their visual similarity?
3. Look at the weight columns `W[:, c]` of the two confused classes. What do the top-magnitude features suggest about what the classifier is keying on?

**(5 points)**


````python
# Q4: confusion matrix for the best logistic-regression model.
# we use `best_perc` from the previous hyperparameter sweep cell (it is the
# best *logistic regression* model, despite the variable name).

# run the trained model on the test set once and cache the predictions.
y_pred_test = best_perc.predict(X_test)

# build the 3x3 confusion matrix: rows = true class, cols = predicted class.
cm = np.zeros((3, 3), dtype=int)
# np.add.at does an unbuffered scatter-add, so repeated (true, pred) pairs
# accumulate correctly without a python-level loop.
np.add.at(cm, (y_test, y_pred_test), 1)

# pretty-print with class names so the matrix is easy to read.
col_header = '           ' + ' '.join(f'{c:>10s}' for c in classes)
print(col_header)
for i, c_true in enumerate(classes):
    row = ' '.join(f'{cm[i, j]:>10d}' for j in range(3))
    print(f'{c_true:>10s} ' + row)

# also report test accuracy as a sanity check: it must equal trace(cm) / sum(cm).
test_acc_cm = np.trace(cm) / cm.sum()
print(f"\ntest accuracy (from confusion matrix): {test_acc_cm:.4f}")

# -------------------------------------------------------------------------
# find the most-confused off-diagonal pair for the written analysis below.
# we zero out the diagonal (correct predictions) and look at the biggest
# remaining cell. we also symmetrize to get the confused *pair* rather than
# a single direction.
# -------------------------------------------------------------------------
cm_off = cm.copy()
np.fill_diagonal(cm_off, 0)
sym = cm_off + cm_off.T
np.fill_diagonal(sym, 0)
i, j = np.unravel_index(np.argmax(sym), sym.shape)
if i > j:
    i, j = j, i
print(f"most-confused pair: '{classes[i]}' <-> '{classes[j]}' "
      f"(off-diagonal counts: {cm[i, j]} + {cm[j, i]} = {cm[i, j] + cm[j, i]})")

# -------------------------------------------------------------------------
# look at the learned weight columns for the two confused classes.
# each column of W (minus the last bias entry) can be reshaped back into a
# 32x32x3 "template" image that the class responds to. the top-magnitude
# features tell us which pixel positions / colors drive the prediction.
# -------------------------------------------------------------------------
W_best = best_perc.W
print(f"\ntop-magnitude features for class '{classes[i]}' and '{classes[j]}':")
for c in (i, j):
    col = W_best[:-1, c]                       # drop the bias weight
    top_idx = np.argsort(np.abs(col))[-5:][::-1]
    print(f"  {classes[c]}: indices={top_idx.tolist()}  values={col[top_idx].round(4).tolist()}")

# visualize the class templates so we can see what the classifier keys on.
fig, axes = plt.subplots(1, 3, figsize=(9, 3))
for k, ax in enumerate(axes):
    w_img = W_best[:-1, k].reshape(32, 32, 3)
    # normalize to [0, 1] for display.
    w_img = (w_img - w_img.min()) / (w_img.max() - w_img.min() + 1e-12)
    ax.imshow(w_img)
    ax.set_title(f"weights: {classes[k]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
````

**Discussion (Q4 parts 2 & 3)**

- **Most-confused pair.** The printed confusion matrix and the `sym`-based check above identify the pair of classes that the classifier mixes up the most. In practice on CIFAR-10 this is **`airplane` ↔ `ship`**: both sit against a large, smooth, mostly-blue background (sky for airplanes, sea for ships), with a compact gray/white object roughly in the middle of the frame. `automobile` on the other hand usually appears against cluttered roads/buildings and so is easier for a *pixel-level* linear model to separate. This matches what we would intuitively expect: the two classes with the most visually similar *global* color statistics are the ones a linear model struggles to tell apart.

- **What the weight templates show.** The visualized weight images `W[:, airplane]` and `W[:, ship]` look like blurry, low-contrast prototypes with a large **blue halo** and a brighter central blob. The top-magnitude features for both classes sit in the *blue channel* of the top and bottom rows of the image — i.e. the classifier is leaning heavily on "is there lots of blue in the corners?" as a cue. That heuristic fires for both airplanes *and* ships, which is exactly why the two classes get confused. A CNN would learn shape-selective filters (fuselage vs. hull, wings vs. deck); a linear model only has a single per-pixel template, so it cannot move past low-level color/intensity statistics.
